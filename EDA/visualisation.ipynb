```python
# Importation des bibliothèques nécessaires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
data = pd.read_csv("../data/HeartDisease.csv")
```

```python
# Paramètres de style pour Seaborn
sns.set(style="whitegrid")
```

# les distributions de chaque variable numérique

```python
plt.figure(figsize=(14, 8))
data.select_dtypes(include=['float64', 'int64']).hist(bins=15, edgecolor='black', color='skyblue', figsize=(14, 8))
plt.tight_layout()
plt.suptitle("Distribution des Variables Numériques", fontsize=16, y=1.02)
plt.show()
```

# 2. Boxplots pour détecter les valeurs extrêmes de chaque variable numérique

```python
plt.figure(figsize=(14, 8))
for i, column in enumerate(data.select_dtypes(include=['float64', 'int64']).columns, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(y=data[column], color='skyblue')
    plt.title(f'Boxplot de {column}')
plt.tight_layout()
plt.suptitle("Boxplots des Variables Numériques", fontsize=16, y=1.02)
plt.show()
```

# 3. Heatmap de corrélation pour visualiser les relations entre les variables numériques

```python
plt.figure(figsize=(10, 8))
correlation_matrix = data.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Heatmap des Corrélations")
plt.show()
```
